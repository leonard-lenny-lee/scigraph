```python
from scigraph.datatables import ColumnTable

data = [
    [3.4, 2.3, 4.2],
    [4.3, 5.2, 7.8],
    [3.0, 4.5, 5.9],
    [3.9, 3.1, 6.4],
    [4.1, 5.0, 7.6],
]

table = ColumnTable(data, ["Untreated", "Placebo", "Treated"])
table.x_title = "Group"
table.y_title = "Response"
table
```

```python
graph = table.create_column_graph("horizontal")
ax = graph \
    .add_box_and_whiskers() \
    .add_points("swarm") \
    .draw()
ax.set_xlim(0, 9.01)
```
