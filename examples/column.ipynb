```python
from scigraph.datatables import ColumnTable

data = [
    [3.4, 2.3, 4.2],
    [4.3, 5.2, 7.8],
    [3.0, 4.5, 5.9],
    [3.9, 3.1, 6.4],
    [4.1, 5.0, 7.6],
]

table = ColumnTable(data, ["Untreated", "Placebo", "Treated"])
table.x_title = "Group"
table.y_title = "Response"
table
```

```python
from scigraph.analyses import WelchTTest

ttest_1 = WelchTTest(table, ("Treated", "Placebo"))
ttest_2 = WelchTTest(table, ("Untreated", "Placebo"))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(2, 3))
graph = table.create_column_graph("vertical")
graph \
    .add_box_and_whiskers() \
    .add_points("swarm") \
    .link_analysis(ttest_1) \
    .link_analysis(ttest_2) \
    .draw(ax=ax);
```
